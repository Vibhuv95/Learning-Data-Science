```python
import numpy as np
```

```python
arr = np.array([1, 2 ,3, 4, 5])
result = arr ** 2 # Vectorized operation
print(result) # Output: [1 4 9 16 25]
```

```python
result + 10
```

```python
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([1, 2, 3])
result = arr1 + arr2 # Brodcasting arr2 across arr1 
print(result)
```

```python
data = np.array([[10, 20, 30],
[15, 25, 35],
[20, 30, 40],
[25, 35, 45],
[30, 40, 50]])
# Calculating mean and standard deviation for each feacher (column)
mean = data.mean(axis = 0)
std = data.std(axis = 0)
# Normalized the data using brodcasting 
normalized_data = (data - mean) / std
print(normalized_data)
```
