```python
import pandas as pd
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_excel("iris-train.xlsx")
```

```python
data.head()
```

```python
x = data.iloc[:, :-1]
```

```python
x.values
```

```python
y = data.iloc[:, -1]
```

```python
y
```

```python
model = GaussianNB()
```

```python
model.fit(x,y)
```

```python
prediction = model.predict([[5.4, 3, 4.5, 1.5]])
```

```python
prediction
```

```python
datatest = pd.read_excel("iris-test.xlsx")
```

```python
x_test = datatest.iloc[:, :-1]
```

```python
prediction = model.predict(x_test.values)
```

```python
prediction
```

```python
y_test = datatest.iloc[:, -1]
```

```python
y_test.values
```

```python
actual = y_test.values
```

```python
count = 0
for i in range(0, len(actual)):
     if actual[i] == prediction[i]:
         count = count + 1
print((count* 100)/ len(actual))    
```
